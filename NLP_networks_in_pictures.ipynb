## LSTM
Below are diagrams from detailed to gross

### [Standard diagram](https://colah.github.io/posts/2015-08-Understanding-LSTMs/)
Every LSTM has two outputs--One is known as '*state*' output or '*cell state*' and the other is simply called *output*. Generally when LSTMs are chained together, the '*cell-state*' output of one feeds another. In the below diagram, of the two (horizontal) outputs, the upper output is the *cell-state* output. <br>
In Sequential model, <i>Cell State</i> output is NOT available. It becomes available if you set *return_state = True* in the syntax. By default, it is set to False. Both outputs are available in Functional Model.

image.png

In an LSTM block we have in all 4-neural networks. Three of the inner NNs have <i>'sigma'</i>. as an activation function and one of them has <i>'tanh'</i> as activation function. Boxed network below is one of the three sigma networks. It has an Input layer and a hidden layer.
